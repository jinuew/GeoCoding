```python
import pandas as pd
from geopy.geocoders import Nominatim
import time

def geocode(address):
    geolocator = Nominatim(user_agent="parkjinwoo")
    location = geolocator.geocode(address)
    if location:
        return (location.longitude, location.latitude)
    else:
        return (None, None)


# CSV 파일 불러오기
csv_path = "path.csv"
data = pd.read_csv(csv_path, encoding='CP949')  # 인코딩 문제 없으면 cp949 빼도 됨
for idx, row in data.iterrows():
    address = row["주소"]  # 컬럼명 맞추기
    if pd.isna(address):
        continue

    longitude, latitude = geocode(address)
    data.at[idx, 'F'] = longitude # 컬럼 설정
    data.at[idx, 'G'] = latitude # 컬럼 설정

    print(f"{address}: 경도 = {longitude}, 위도 = {latitude}")

    time.sleep(1)  # API 요청 제한을 위한 지연 / 뭔지 잘 모르겠

# 결과 파일 생성
data.to_csv("set_file_name.csv", index=False) #경로에 저장 됨
print("!")
```
