```python
# 파일 업로드형

import requests
import json
import pandas as pd

def Kakao_Geocoding(address, api_key):
    url = 'https://dapi.kakao.com/v2/local/search/address.json'
    headers = {"Authorization": f"KakaoAK YOUR_KEY"}
    params = {'query': address}
    response = requests.get(url, headers=headers, params=params)
    return response.json()


api_key = 'YOUR_KEY'

# read CSV
csv_file_path = 'path.csv'
addresses_df = pd.read_csv(csv_file_path, encoding='CP949')

# 리스트 생성
latitudes = []
longitudes = []

# '주소' 열의 각 주소에 대해 Kakao Geocoding API를 호출
for address in addresses_df['주소']: #컬렁명 일치
    try:
        result = Kakao_Geocoding(address.strip(), api_key)
        address_info = result['documents'][0]['address']
        latitudes.append(address_info['y'])
        longitudes.append(address_info['x'])
    except (IndexError, KeyError):
        # 에러 -> None
        latitudes.append(None)
        longitudes.append(None)

# 위경도 데이터프레임 추가
addresses_df['latitude'] = latitudes
addresses_df['longitude'] = longitudes

# 완성된 데이터프레임을 새 CSV 파일로 저장합니다.
output_csv_path = 'set_file_name.csv'
addresses_df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')
```
